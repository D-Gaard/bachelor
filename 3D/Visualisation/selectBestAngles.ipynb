```python
p = r"3dTo2d\optimalView.PNG"
img = cv2.imread(p)
#img = cv2.bitwise_not(img)
img = img[..., ::-1]
img_r = cv2.resize(img, [61, 181], fx=0.5, fy=0.5)
#img_r = cv2.cvtColor(img_r, cv2.COLOR_BGR2RGB)
```

```python
sel_angles = [(-87,27),(-75,-22),(0,0),(87,-27),(87,27),(-43,13),(-43,-11),(43,-13),(43,13)]
x = []
y = []
for i in range(len(sel_angles)):
  prim, sec = sel_angles[i]
  x.append(sec)
  y.append(prim)

im = plt.imread(p)
im = img_r
plt.figure(figsize=(8, 10))
implot = plt.imshow(
    im, extent=[-im.shape[1]/2., im.shape[1]/2., -im.shape[0]/2., im.shape[0]/2.], origin='lower')
plt.scatter(x,y,c="gray")
plt.xlim(-30, 30)
plt.ylim(90, -90)
plt.xlabel("Secondary angle")
plt.ylabel("Primary angle")

for i, txt in enumerate(sel_angles):
    plt.annotate(txt, (x[i], y[i]),c="black")

plt.show()
```
